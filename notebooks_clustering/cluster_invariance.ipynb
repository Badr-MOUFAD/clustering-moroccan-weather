```python
# imports
import numpy as np
import pandas as pd
```

```python
# imports for plots
from plotly import graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

pio.templates.default = "plotly_white"
```

```python
# load data with const latitude
import os

dir_name = "data_inv_lat"
df_const_lat = pd.DataFrame(columns=["lat", "lon", "year"," cluster"])

for filename in os.listdir(dir_name):
    path = os.path.join(dir_name, filename)
    df_const_lat = df_const_lat.append(pd.read_csv(path), ignore_index=True)
```

```python
# remove null col
df_const_lat = df_const_lat.dropna(axis=1)

df_const_lat
```

```python
# load data with const lon
import os

dir_name = "data_inv_lon"
df_const_lon = pd.DataFrame(columns=["lat", "lon", "year"," cluster"])

for filename in os.listdir(dir_name):
    path = os.path.join(dir_name, filename)
    df_const_lon = df_const_lon.append(pd.read_csv(path), ignore_index=True)
```

```python
# remove null col
df_const_lon = df_const_lon.dropna(axis=1)

df_const_lon
```

```python
# make a spatio-temporal plot

fig = px.scatter(df_const_lat, x="year", y="lon", color="cluster", color_discrete_sequence=['#636EFA', "#00CC96",'#FFA15A'])


fig.update_traces(
    marker_size=22,
    selector=dict(mode='markers')
)

fig.update_layout(
    title="Spatio-temporal map for regions with Latitude=-5.924",
    xaxis_title="crop year",
    yaxis_title="Longitude"
)

fig.show()
```

```python
# make a spatio-temporal plot

fig = px.scatter(df_const_lon, x="year", y="lat", color="cluster", color_discrete_sequence=['#636EFA', "#00CC96",'#FFA15A'])


fig.update_traces(
    marker_size=22,
    selector=dict(mode='markers')
)

fig.update_layout(
    title="Spatio-temporal map for regions with Longitude=-34.2623",
    yaxis_title="Longitude",
    xaxis_title="crop year",
    height=370
)

fig.show()
```
